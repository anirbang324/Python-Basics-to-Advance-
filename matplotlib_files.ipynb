```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x and y values for the plot
x = np.array([2, 3, 4])
y = np.array([1, 2, 3])

# Plot the data points with markers
plt.plot(x, y,marker = 'D', linestyle = 'dashdot', c = 'g',lw = 1)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x and y values for the plot

y = np.array([1,3,5,7])

plt.plot(y,marker = 'o')
plt.grid(axis = 'y', color = 'r', lw = 1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x and y values for the plot

y = np.array([3,10,4,1])
plt.subplot(2,1,1)
plt.plot(y,marker = 'o')
plt.grid(axis = 'y', color = 'r', lw = 0.5)
plt.show()

y = np.array([7,1,8,3])
plt.subplot(2,1,2)
plt.plot(y,marker = 'o')
plt.grid(axis = 'y', color = 'g', lw = 0.5)
plt.show()
```
