```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["num1","num2","num3","num4"])
y = np.array([5,1,9,3])

plt.bar(x,y,width = 0.3, color = "red")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2005, 2007, 2009, 2011, 2013])
y = np.array([5, 7, 2, 10, 13])
plt.barh(x, y, height = 0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(150,10,240)
plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,5,9,3,6])
mylabels = ["blue","yellow","red","green","purple"]
myexplode = [0.2,0,0,0,0]

plt.pie(x, labels = mylabels, explode = myexplode, shadow = True)
plt.legend(title = "colors")
plt.show()
```
